```python
import os
import glob

# Check if the directory exists
audio_dir = "../dataset/8sidor_audios"
print(f"Directory exists: {os.path.exists(audio_dir)}")
print(f"Directory contents: {os.listdir(audio_dir) if os.path.exists(audio_dir) else 'Directory not found'}")

# Check for audio files
audio_extensions = ['*.wav', '*.mp3', '*.flac', '*.m4a', '*.ogg']
for ext in audio_extensions:
    files = glob.glob(os.path.join(audio_dir, '**', ext), recursive=True)
    print(f"{ext} files found: {len(files)}")
    if files:
        print(f"  Sample: {files[0]}")
```

```python
import os
import webdataset as wds

# mapp med wav-filer
wav_folder = "../dataset/8sidor_audios"


# samla alla wav-filer
wav_files = [f for f in os.listdir(wav_folder) if f.endswith(".wav")]
wav_files.sort()  # valfritt, för ordning

os.makedirs(os.path.dirname("../dataset/8sidor_audios_dataset/"), exist_ok=True)
# skapa TAR
with wds.ShardWriter("../dataset/8sidor_audios_dataset/dataset-%06d.tar", maxsize=5e9) as sink: # type: ignore
    for i, wav_file in enumerate(wav_files):
        path = os.path.join(wav_folder, wav_file)
        with open(path, "rb") as f:
            sample = {
                "__key__": str(i).zfill(6),  # nyckel för varje exempel
                ".wav": f.read()             # innehållet
            }
            sink.write(sample)
```
