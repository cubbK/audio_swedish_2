```python
import librosa
import os
from pathlib import Path

def get_total_audio_length_librosa(folder_path):
    """Get total duration of all WAV files in a folder using librosa"""
    total_duration = 0
    wav_files = []
    
    # Get all WAV files in the folder
    for file_path in Path(folder_path).glob("*.mp3"):
        try:
            # Load audio file and get duration
            duration = librosa.get_duration(path=str(file_path))
            total_duration += duration
            wav_files.append((file_path.name, duration))
            print(f"{file_path.name}: {duration:.2f} seconds")
        except Exception as e:
            print(f"Error processing {file_path.name}: {e}")
    
    print(f"\nTotal files: {len(wav_files)}")
    print(f"Total duration: {total_duration:.2f} seconds ({total_duration/60:.2f} minutes)")
    return total_duration, wav_files

# Usage
# dataset/8_sidor_audios_full
folder_path = "../dataset/8_sidor_audios_full"
total_duration, files_info = get_total_audio_length_librosa(folder_path)

print(total_duration, files_info)

print(f"Total durations in hours: {total_duration / 3600:.2f} hours")
```
