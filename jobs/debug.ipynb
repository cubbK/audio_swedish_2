```python
import tarfile
import os

def create_tar_archive():
    with tarfile.open(f"test.tar", "w") as tar:
        for root, dirs, files in os.walk("audio_segments"):
            for file in files:
                file_path = os.path.join(root, file)
                tar.add(file_path, arcname=file)
                
create_tar_archive()
```
